```python
# Import necessary libraries
import pandas as pd
```

```python
# Load the Dataset into pandas data frame
data = pd.read_csv("pll-team-stats.csv")
```

```python
#To get quick info of data
data.info()

# to get top 5 values 
data.head()
```

```python
# Display information about missing values in the data
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Display initial statistics
# Note that data.describe() only provides statistics for numeric columns. 
# If you want to include all columns in the summary statistics, you can use data.describe(include='all').

statistics = data.describe(include='all')
print(statistics)
```

```python
# Check the dimensions of the data frame -> gives no.of rows and no.of columns
dimensions = data.shape
print(dimensions)
```

```python
# Summarize the types of variables by checking the data types and apply proper type conversions
data_types = data.dtypes
print(data_types)
```

```python
# If you need to convert variables to a specific data type:
# Assuming 'column_name' needs to be converted to a numeric type

data['penalty_minutes'] = pd.to_numeric(data['penalty_minutes'])
```

```python
print(data_types)
```

```python
# Turn categorical variables into quantitative variables using one hot encoding
# Assuming 'categorical_column' is the column you want to encode

one_hot_encoded_data = pd.get_dummies(data, columns=['team'])
print(one_hot_encoded_data)
```
