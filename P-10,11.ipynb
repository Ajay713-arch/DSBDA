```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset into a pandas DataFrame
df = pd.read_csv('ds_salaries.csv')
```

```python
# Step 2: Get an overview of features and their data types
print(df.info())
```

```python
feature_info = df.dtypes
print(feature_info)
```

```python
# Step 3: Create a histogram for each feature
df.hist()
plt.show()
```

```python
# Create histograms for all features
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# alpha is used to control transperacy of bar
# histogram used for numeric variable

# Specify the variables for the histogram
x = df['work_year']
y = df['salary']

# Create a histogram
plt.hist(x, bins=10, alpha=0.5, label='Work Year')
plt.hist(y, bins=10, alpha=0.5, label='Salary')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Show the histogram
plt.show()
```

```python
# Step 4: Create a box plot for each feature
df.boxplot()
plt.show()
```

```python
# Create a box plot between two variables
sns.boxplot(x='work_year', y='salary', data=df)
plt.show()
```
