```python
# Performing min-max normalization on oulier free data
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()
```

```python
# Min-max normalizaation can be applied to only numeric data set hence first convert all into quantative dataset

# Identify columns with non-numeric values
# non_numeric_columns = df.select_dtypes(exclude=['float', 'int']).columns
# print(non_numeric_columns)

# from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
# encoder = LabelEncoder()

# Encode categorical variables
#for column in non_numeric_columns:
 #   df[column] = encoder.fit_transform(df[column])
```

```python
# Select the column you want to normalize
column_to_normalize = df['Height']

# Reshape the column to a 2-dimensional array (required by the scaler)
reshaped_column = column_to_normalize.values.reshape(-1, 1)

# Apply min-max normalization to the column
normalized_column = scaler.fit_transform(reshaped_column)

# Replace the original column with the normalized values
df['Height'] = normalized_column
```
